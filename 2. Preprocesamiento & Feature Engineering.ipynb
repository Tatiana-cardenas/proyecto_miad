```python
def resumen_general_df(nombre_df, df):
    filas, columnas = df.shape
    
    duplicados = df.duplicated().sum()
    
    total_nulos = df.isna().sum().sum()
    
    # Cuenta celdas vacías tipo "", "   " solo en columnas object/string
    total_vacios = (
        df.select_dtypes(include=["object", "string"])
        .apply(lambda col: col.astype(str).str.strip().eq("").sum())
        .sum()
    )
    
    columnas_con_nulos = df.columns[df.isna().any()].tolist()
    
    columnas_con_vacios = []
    for col in df.select_dtypes(include=["object", "string"]).columns:
        if df[col].astype(str).str.strip().eq("").any():
            columnas_con_vacios.append(col)
    
    return pd.DataFrame({
        "DataFrame": [nombre_df],
        "Filas": [filas],
        "Columnas": [columnas],
        "Duplicados": [duplicados],
        "Total nulos": [total_nulos],
        "Total vacíos": [total_vacios],
        "Columnas con nulos": [", ".join(columnas_con_nulos) if columnas_con_nulos else "Ninguna"],
        "Columnas con vacíos": [", ".join(columnas_con_vacios) if columnas_con_vacios else "Ninguna"]
    })
```

```python
def limpiar_texto(x):
    if pd.isna(x):
        return np.nan
    
    x = str(x).strip().upper()
    
    # Quitar tildes
    x = ''.join(
        c for c in unicodedata.normalize('NFD', x)
        if unicodedata.category(c) != 'Mn'
    )
    
    return x
```

```python
# =========================
# 3. Feature engineering térmico
# =========================

df_fe["amplitud_temperatura_media_mensual"] = (
    df_fe["máximo de la temperatura media mensual (°c)"]
    - df_fe["mínimo de la temperatura media mensual (°c)"]
)

# =========================
# 4. Feature engineering hídrico
# =========================

df_fe["balance_precipitacion_evaporacion"] = (
    df_fe["precipitación acumulada anual (mm/año)"]
    - df_fe["evaporación potencial acumulada anual (mm/año)"]
)

df_fe["ratio_precipitacion_evaporacion"] = (
    df_fe["precipitación acumulada anual (mm/año)"]
    / (df_fe["evaporación potencial acumulada anual (mm/año)"] + 1e-6)
)

# Transformaciones logarítmicas para acumulados positivos
cols_log = [
    "área sembrada",
    "precipitación acumulada anual (mm/año)",
    "radiación solar acumulada anual (mj/m²/año)",
    "evaporación potencial acumulada anual (mm/año)"
]

for col in cols_log:
    if col in df_fe.columns:
        df_fe[f"log1p_{col}"] = np.log1p(df_fe[col])

# =========================
# 5. Humedad del suelo
# =========================

capa1 = "humedad volumétrica media anual del suelo capa 1 (m³/m³)"
capa2 = "humedad volumétrica media anual del suelo capa 2 (m³/m³)"

df_fe["humedad_suelo_promedio_capas_1_2"] = (
    df_fe[capa1] + df_fe[capa2]
) / 2

df_fe["humedad_suelo_diferencia_capa2_capa1"] = (
    df_fe[capa2] - df_fe[capa1]
)

# =========================
# 6. SPI: intensidad y duración de sequía
# =========================

for escala in [3, 6, 12]:
    col_min = f"spi{escala}_min_anual"
    col_meses = f"spi{escala}_meses_bajo_m1"

    df_fe[f"spi{escala}_deficit_max_anual"] = np.maximum(
        0,
        -df_fe[col_min]
    )

    df_fe[f"spi{escala}_presion_sequia_anual"] = (
        df_fe[f"spi{escala}_deficit_max_anual"] * df_fe[col_meses]
    )

df_fe["spi_presion_total_sequia"] = (
    df_fe["spi3_presion_sequia_anual"]
    + df_fe["spi6_presion_sequia_anual"]
    + df_fe["spi12_presion_sequia_anual"]
)

df_fe["spi_promedio_dic"] = (
    df_fe["spi3_dic"] + df_fe["spi6_dic"] + df_fe["spi12_dic"]
) / 3

df_fe["spi1_cambio_llenado_vs_floracion"] = (
    df_fe["spi1_llenado"] - df_fe["spi1_floracion"]
)

df_fe["spi3_cambio_llenado_vs_floracion"] = (
    df_fe["spi3_llenado"] - df_fe["spi3_floracion"]
)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
TARGET = "rendimiento"

# Variables numéricas predictoras, excluyendo target
variables_numericas_modelo = [
    col for col in df_fe.select_dtypes(include=[np.number]).columns
    if col != TARGET
]

X_num = df_fe[variables_numericas_modelo].copy()

# Estandarizar
scaler = StandardScaler()
Z = scaler.fit_transform(X_num)

# PCA sin target
pca = PCA()
scores = pca.fit_transform(Z)

componentes = [f"PC{i}" for i in range(1, len(variables_numericas_modelo) + 1)]

loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

loadings_df = pd.DataFrame(
    loadings,
    index=variables_numericas_modelo,
    columns=componentes
)

pca_resumen = pd.DataFrame({
    "componente": componentes,
    "autovalor": pca.explained_variance_,
    "varianza_explicada": pca.explained_variance_ratio_,
    "varianza_acumulada": np.cumsum(pca.explained_variance_ratio_)
})

pca_resumen.head(15)
```

```python
# Usar componentes que expliquen mínimo 80% de la varianza
umbral_varianza = 0.90

n_componentes = np.argmax(
    pca_resumen["varianza_acumulada"] >= umbral_varianza
) + 1

pcs_usar = [f"PC{i}" for i in range(1, n_componentes + 1)]

# Score PCA: carga absoluta promedio en los componentes importantes
pca_importancia = pd.DataFrame({
    "variable": loadings_df.index,
    "score_pca": loadings_df[pcs_usar].abs().mean(axis=1),
    "max_loading_abs": loadings_df[pcs_usar].abs().max(axis=1)
}).sort_values("score_pca", ascending=False)

pca_importancia.head(20)
```

```python
corr_target = []

for col in variables_numericas_modelo:
    pearson = df_fe[[col, TARGET]].corr(method="pearson").iloc[0, 1]
    spearman = df_fe[[col, TARGET]].corr(method="spearman").iloc[0, 1]
    
    corr_target.append({
        "variable": col,
        "pearson_r_target": pearson,
        "spearman_r_target": spearman,
        "abs_pearson_target": abs(pearson),
        "abs_spearman_target": abs(spearman),
        "score_corr_target": max(abs(pearson), abs(spearman))
    })

corr_target_df = pd.DataFrame(corr_target).sort_values(
    "score_corr_target",
    ascending=False
)

corr_target_df.head(20)
```

```python
def calcular_vif(X):
    X_temp = X.copy()
    X_temp = X_temp.replace([np.inf, -np.inf], np.nan)
    X_temp = X_temp.fillna(X_temp.median())

    X_const = sm.add_constant(X_temp, has_constant="add")

    resultados = []

    for i, col in enumerate(X_const.columns):
        if col == "const":
            continue

        vif = variance_inflation_factor(X_const.values, i)

        resultados.append({
            "variable": col,
            "VIF": vif
        })

    return pd.DataFrame(resultados).sort_values("VIF", ascending=False)


vif_df = calcular_vif(X_num)

vif_df.head(30)
```

```python
resumen_metricas = (
    corr_target_df
    .merge(pca_importancia, on="variable", how="left")
    .merge(vif_df, on="variable", how="left")
)

resumen_metricas["VIF_limpio"] = resumen_metricas["VIF"].replace(np.inf, 1e9)

resumen_metricas.sort_values("VIF_limpio", ascending=False).head(30)
```

```python
def seleccionar_variables_por_grupo(resumen_metricas, grupos_redundantes):
    eliminadas = []
    mantenidas = []

    for grupo in grupos_redundantes:
        nombre_grupo = grupo["grupo"]
        variables = [v for v in grupo["variables"] if v in resumen_metricas["variable"].values]
        max_keep = grupo["max_keep"]

        if len(variables) <= max_keep:
            continue

        temp = resumen_metricas[resumen_metricas["variable"].isin(variables)].copy()

        # Mejor variable:
        # 1. mayor correlación con target
        # 2. mayor importancia PCA
        # 3. menor VIF
        temp = temp.sort_values(
            by=["score_corr_target", "score_pca", "VIF_limpio"],
            ascending=[False, False, True]
        )

        vars_keep = temp.head(max_keep)["variable"].tolist()
        vars_drop = temp.iloc[max_keep:]["variable"].tolist()

        for v in vars_keep:
            mantenidas.append({
                "grupo": nombre_grupo,
                "variable": v,
                "decision": "mantener"
            })

        for v in vars_drop:
            eliminadas.append({
                "grupo": nombre_grupo,
                "variable": v,
                "decision": "eliminar",
                "razon": "Redundante dentro del grupo; menor correlación con target, menor PCA o mayor VIF"
            })

    return pd.DataFrame(mantenidas), pd.DataFrame(eliminadas)


mantenidas_grupos, eliminadas_grupos = seleccionar_variables_por_grupo(
    resumen_metricas,
    grupos_redundantes
)

eliminadas_grupos
```

```python
def pares_altamente_correlacionados(corr_matrix, umbral=0.85):
    pares = []

    cols = corr_matrix.columns

    for i in range(len(cols)):
        for j in range(i + 1, len(cols)):
            r = corr_matrix.iloc[i, j]

            if abs(r) >= umbral:
                pares.append({
                    "variable_1": cols[i],
                    "variable_2": cols[j],
                    "correlacion": r,
                    "abs_correlacion": abs(r)
                })

    return pd.DataFrame(pares).sort_values("abs_correlacion", ascending=False)


corr_predictoras = X_num.corr(method="pearson")

pares_corr_alta = pares_altamente_correlacionados(
    corr_predictoras,
    umbral=0.85
)

pares_corr_alta.head(30)
```

```python
def decidir_eliminar_por_pares(pares_corr_alta, resumen_metricas, variables_ya_eliminadas=None):
    if variables_ya_eliminadas is None:
        variables_ya_eliminadas = set()

    eliminadas = []

    metricas = resumen_metricas.set_index("variable")

    for _, row in pares_corr_alta.iterrows():
        v1 = row["variable_1"]
        v2 = row["variable_2"]

        if v1 in variables_ya_eliminadas or v2 in variables_ya_eliminadas:
            continue

        if v1 not in metricas.index or v2 not in metricas.index:
            continue

        m1 = metricas.loc[v1]
        m2 = metricas.loc[v2]

        # Regla:
        # eliminar la que tenga menor correlación con target.
        # si empatan, eliminar mayor VIF.
        # si empatan, eliminar menor PCA.
        score1 = (
            m1["score_corr_target"],
            m1["score_pca"],
            -m1["VIF_limpio"]
        )

        score2 = (
            m2["score_corr_target"],
            m2["score_pca"],
            -m2["VIF_limpio"]
        )

        if score1 >= score2:
            eliminar = v2
            mantener = v1
        else:
            eliminar = v1
            mantener = v2

        variables_ya_eliminadas.add(eliminar)

        eliminadas.append({
            "variable_eliminada": eliminar,
            "variable_mantenida": mantener,
            "correlacion": row["correlacion"],
            "abs_correlacion": row["abs_correlacion"],
            "razon": "Alta correlación entre predictoras"
        })

    return pd.DataFrame(eliminadas)


variables_eliminadas_iniciales = set(eliminadas_grupos["variable"].tolist())

eliminadas_corr = decidir_eliminar_por_pares(
    pares_corr_alta,
    resumen_metricas,
    variables_ya_eliminadas=variables_eliminadas_iniciales
)

eliminadas_corr.head(30)
```

```python
def reducir_vif_iterativo(X, resumen_metricas, variables_excluir=None, umbral_vif=10):
    if variables_excluir is None:
        variables_excluir = set()

    variables_actuales = [
        col for col in X.columns
        if col not in variables_excluir
    ]

    historial = []

    while True:
        vif_actual = calcular_vif(X[variables_actuales])
        vif_actual["VIF_limpio"] = vif_actual["VIF"].replace(np.inf, 1e9)

        max_vif = vif_actual["VIF_limpio"].max()

        if max_vif <= umbral_vif or len(variables_actuales) <= 2:
            break

        candidatas = vif_actual[vif_actual["VIF_limpio"] > umbral_vif].copy()

        temp = candidatas.merge(
            resumen_metricas[["variable", "score_corr_target", "score_pca"]],
            on="variable",
            how="left"
        )

        # Eliminar la peor:
        # menor relación con target,
        # menor importancia PCA,
        # mayor VIF
        temp = temp.sort_values(
            by=["score_corr_target", "score_pca", "VIF_limpio"],
            ascending=[True, True, False]
        )

        var_eliminar = temp.iloc[0]["variable"]

        historial.append({
            "variable_eliminada": var_eliminar,
            "VIF": temp.iloc[0]["VIF"],
            "score_corr_target": temp.iloc[0]["score_corr_target"],
            "score_pca": temp.iloc[0]["score_pca"],
            "razon": f"VIF mayor a {umbral_vif}"
        })

        variables_actuales.remove(var_eliminar)

    vif_final = calcular_vif(X[variables_actuales])

    return variables_actuales, pd.DataFrame(historial), vif_final


variables_excluir = set(eliminadas_grupos["variable"].tolist()) | set(eliminadas_corr["variable_eliminada"].tolist())

variables_finales_lineal, eliminadas_vif, vif_final = reducir_vif_iterativo(
    X=X_num,
    resumen_metricas=resumen_metricas,
    variables_excluir=variables_excluir,
    umbral_vif=10
)

eliminadas_vif
```
